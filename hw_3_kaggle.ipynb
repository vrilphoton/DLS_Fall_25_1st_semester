# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline

from sklearn.compose import ColumnTransformer

from catboost import CatBoostClassifier
```

```python
y = data[target_col]
X = data.drop(columns = [target_col])

X_train, X_valid, y_train, y_valid = train_test_split(X,y, test_size = 0.25, random_state = 42)

preprocessor = ColumnTransformer (
    transformers = [
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(drop = 'first'), cat_cols)
    ],
    remainder = 'passthrough'
)

pipe = make_pipeline(preprocessor, LogisticRegressionCV(Cs = [100, 10, 1, 0.1, 0.01, 0.001], cv = 5, scoring = 'roc_auc', refit = True))

pipe.fit(X_train, y_train)
best_C = pipe.named_steps['logisticregressioncv'].C_[0]
print(f'Best_C : {best_C}')
best_score = pipe.named_steps['logisticregressioncv'].scores_[1].max()
print(f'Best ROC-AUC scoring: {best_score:.4f}')
```

```python
from sklearn.metrics import roc_auc_score


y_pred = pipe.predict(X_valid)

print(f'ROC-AUC for _valid: {roc_auc_score(y_pred, y_valid)}')
```

## Применение градиентного бустинга (2 балла)

Если вы хотите получить баллы за точный ответ, то стоит попробовать градиентный бустинг. Часто градиентный бустинг с параметрами по умолчанию даст вам 80% результата за 0% усилий.

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

2) Протестируйте разные занчения параметров количества деревьев и learning_rate'а и выберите лучшую по метрике ROC-AUC комбинацию.

(Дополнительно) Есть некоторые сложности с тем, чтобы использовать CatBoostClassifier вместе с GridSearchCV, поэтому мы не просим использовать кроссвалидацию. Но можете попробовать)

```python
catboost_1 = CatBoostClassifier(
    iterations=500,
    learning_rate=0.1,
    depth=6,
    loss_function='Logloss',
    eval_metric='AUC',
    random_seed=42,
    verbose=100,
    cat_features=cat_cols
)

catboost_1.fit(X_train, y_train, eval_set = (X_valid, y_valid), use_best_model = True)
```

```python
catboost_2 = CatBoostClassifier(
    iterations=100,
    learning_rate=0.1,
    depth=6,
    loss_function='Logloss',
    eval_metric='AUC',
    random_seed=42,
    verbose=100,
    cat_features=cat_cols
)

catboost_2.fit(X_train, y_train, eval_set = (X_valid, y_valid), use_best_model = True)
```

```python
catboost_3 = CatBoostClassifier(
    iterations=500,
    learning_rate=0.2,
    depth=6,
    loss_function='Logloss',
    eval_metric='AUC',
    random_seed=42,
    verbose=100,
    cat_features=cat_cols
)

catboost_3.fit(X_train, y_train, eval_set = (X_valid, y_valid), use_best_model = True)
```

```python
Models = [catboost_1, catboost_2, catboost_3]
```
